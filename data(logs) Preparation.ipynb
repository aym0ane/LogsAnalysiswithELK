#### Function that transforms the lignes of log.txt into a dictionary of key:value

```python
def dict1_log(lines):
    for line in lines:
        if (line.startswith('starttime: ')):
            starttime = re.search('starttime: (.*)', line).group(1)
        elif (line.startswith('master: ')):
            master = re.search('master: (.*)',line).group(1)
        elif (line.startswith('results: ')):
            results = re.search('results: (.*)', line).group(1)
        elif ('     INFO -     "project"' in line):
            project = re.search('"project": "(.*)",', line).group(1)
        elif('     INFO -     "product"' in line):
            product = re.search('"product": "(.*)",', line).group(1)
        elif('     INFO -     "script_repo_revision"' in line):
            script_repo_revision = re.search('"script_repo_revision": "(.*)",', line).group(1)
        elif('     INFO -     "scheduler"' in line):
            scheduler = re.search('"scheduler": "(.*)",', line).group(1)
        elif('     INFO -     "repository"' in line):
            repository = re.search('"repository": "(.*)",', line).group(1)
        elif('     INFO -     "buildername"' in line):
            buildername = re.search('"buildername": "(.*)",', line).group(1)
        elif('     INFO -     "buildid"' in line):
            buildid = re.search('"buildid": "(.*)",', line).group(1)
        elif('     INFO -     "pgo_build"' in line):
            pgo_build = re.search('"pgo_build": "(.*)",', line).group(1)
        elif('     INFO -     "basedir"' in line):
            basedir = re.search('"basedir": "(.*)",', line).group(1)
        elif('     INFO -     "buildnumber"' in line):
            buildnumber = re.search('"buildnumber": (.*),', line).group(1)
        elif('     INFO -     "slavename"' in line):
            slavename = re.search('"slavename": "(.*)",', line).group(1)
            taskos = get_os_name(slavename)
        elif('     INFO -     "revision"' in line):
            revision = re.search('"revision": "(.*)",', line).group(1)
        elif('     INFO -     "master"' in line):
            master = re.search('"master": "(.*)",', line).group(1)
        elif('     INFO -     "platform"' in line):
            platform = re.search('"platform": "(.*)",', line).group(1)
        elif('     INFO -     "branch"' in line):
            branch = re.search('"branch": "(.*)",', line).group(1)
        elif('     INFO -     "repo_path"' in line):
            repo_path = re.search('"repo_path": "(.*)",', line).group(1)
        elif('     INFO -     "moz_repo_path"' in line):
            moz_repo_path = re.search('"moz_repo_path": "(.*)",', line).group(1)
        elif('     INFO -     "stage_platform"' in line):
            stage_platform = re.search('"stage_platform": "(.*)",', line).group(1)
        elif('     INFO -     "builduid"' in line):
            builduid = re.search('"builduid": "(.*)",', line).group(1)
        elif('     INFO -     "slavebuilddir"' in line):
            slavebuilddir = re.search('"slavebuilddir": "(.*)"', line).group(1)
    return {'starttime': float(starttime), 'results': results,'project': project, 'product': product, 'script_repo_revision': script_repo_revision,
                'scheduler': str(scheduler), 'repository': repository, 'buildername': buildername,
               'buildid': buildid, 'pgo_build': pgo_build, 'basedir': basedir, 'buildnumber': int(buildnumber),
               'slavename': slavename, 'taskos': taskos, 'revision': revision, 'master': master, 'platform': platform,
               'branch': branch, 'repo_path': repo_path, 'moz_repo_path': moz_repo_path, 'stage_platform': stage_platform,
               'builduid': builduid, 'slavebuilddir': slavebuilddir}
```

#### function that returs a list of data for each step or block in a log

```python
#Function qui retourne une liste des données des steps/blocks 
def get_log_steps(lines):
    step = []
    j=0
    args = ''
    steps_count = 0
    for i in range(len(lines)):
        if re.search("========= (Started) (.*?)",lines[i]) != None:
            steps_count = 0
            stepname = re.search('========= (Started) (.*?)results',lines[i]).group(2)[:-2]
            j=i
            time1 = datetime.datetime.strptime(re.search('\d+-\d+-\d+ \d+:\d+:\d+\.\d+', lines[i]).group(0), '%Y-%m-%d %H:%M:%S.%f')
            startedDateTime = re.search('at (.*) ', lines[i]).group(1)[:-1]
        elif (re.search("========= (Finished|Finished) (.*?)",lines[i]) == None) and (i>j) and (re.search('argv: ',lines[i]) == None):
            steps_count +=1
        elif (re.search('argv: ',lines[i]) != None) and (i>j) and (re.search("========= (Finished|Finished) (.*?)",lines[i]) == None):
            args = lines[i][lines[i].find("[")+1:lines[i].find("]")].replace("'","").rsplit(', ')
            steps_count +=1
        elif (re.search("========= (Finished) (.*?)",lines[i]) != None) and (re.search('========= (Finished) (.*?)results',lines[i]).group(2)[:-2] == stepname):
            time2 =datetime.datetime.strptime(re.search('\d+-\d+-\d+ \d+:\d+:\d+\.\d+', lines[i]).group(0), '%Y-%m-%d %H:%M:%S.%f')
            endDateTime = re.search('at (.*) ', lines[i]).group(1)[:-1]
            step.append({'stepname': stepname.replace("'",''),'startedDateTime':startedDateTime, 'endDateTime':endDateTime , 'secs': (time2-time1).total_seconds(),'steps_nub':steps_count,'arguments':args})
    return step
```

#### Grouping the logs of a day in a json file 

```python
# Script du code qui retourne un fichier json qui regroupe les logs d'une journée
path = 'C:/Users/desktop/elk/test/'
filename = 'log-2018-06-08'
data =  []
#a=''
final = ''
file_location = os.path.join(path+filename, '*.txt')
filenames = glob.glob(file_location)
print(len(filenames))
for f in filenames:
    logid = re.sub(path+filename+r'\\','',f)
    logid=re.sub("_(\d+)-(\d+)-(\d+)-(\d+)-(\d+)-(\d+)", '',logid)
    logid = logid[:-4]
    with open(f, encoding="utf8") as ff:
        try:
            a0 = {'logid': logid}
            lines = ff.readlines()
            a=dict1_log(lines)
            a['Steps'] = get_log_steps(lines)
            a = dict(a0,**a)
        except:
            pass
            #print(f)
    final +=  str(a).replace("'",'"') + '\n'

    
with open(filename+'.json', 'w', encoding='utf8') as f:
    f.write(final)
```
